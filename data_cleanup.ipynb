```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
calendar = "calendar.csv"
listings = "listings.csv"
reviews = "reviews.csv"
```

```python
calendar_data = pd.read_csv(calendar)
listings_data = pd.read_csv(listings)
reviews_data = pd.read_csv(reviews)

listings_data.head()
```

```python
listings_df = listings_data[['id', 'neighbourhood_cleansed', 'room_type',
                            'zipcode','latitude','longitude', 'property_type', 'price','number_of_reviews',
                            'review_scores_rating', 'reviews_per_month', 'square_feet']]

listings_df.head()
```

```python
listings_df['price'] = pd.to_numeric(listings_df['price'].apply(lambda x:x.replace('$', '').replace(',','')))
```

```python
listings_df['potential_annual_revenue']= listings_df['price']*365*.33
```

```python
zipcode_values = pd.read_csv('HV_zipcode.csv', encoding='latin1')

values = zipcode_values[['RegionName', 'Zhvi']]
values = values.rename(columns={'RegionName':'zipcode', 'Zhvi': 'Zillow_Avg_Value'})
values.head()
```

```python
merged_df = pd.merge(listings_df, values, how='left', on=['zipcode', 'zipcode'])
merged_df.head()
```

```python
merged_df['yield']= merged_df['potential_annual_revenue'] / merged_df['Zillow_Avg_Value'] *100
merged_df['rent_squre_foot']= merged_df['potential_annual_revenue'] / merged_df['square_feet']
merged_df['rent_squre_foot'] = merged_df['rent_squre_foot'].apply(lambda x: round(x,2))

merged_df.head()
```

```python
merged_df = merged_df.fillna(value=0)
merged_df.head()
```

```python
merged_df.to_csv('cleaned_data.csv')
```
