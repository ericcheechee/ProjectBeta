```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

```

```python
calendar = "calendar.csv"
listings = "listings.csv"
reviews = "reviews.csv"
```

```python
calendar_data = pd.read_csv(calendar)
listings_data = pd.read_csv(listings)
reviews_data = pd.read_csv(reviews)

listings_data.head()
```

```python
listings_df = listings_data[['id', 'neighbourhood_cleansed',
                            'zipcode','latitude','longitude', 'property_type', 'price','number_of_reviews',
                            'review_scores_rating', 'reviews_per_month']]

listings_df.head()
```

```python
listings_df['price'] = pd.to_numeric(listings_df['price'].apply(lambda x:x.replace('$', '').replace(',','')))
listings_df.head()
```

```python
listings_df.head()

```

```python
listings_df[['review_scores_rating', 'reviews_per_month']].fillna(value=0)
```

```python
listings_df = listings_df.dropna(axis=0)
```

```python
listings_df.to_csv('cleaned_listing.csv')
```
